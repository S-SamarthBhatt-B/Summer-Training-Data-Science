###Answer 1.

Filter out non-candidate or irrelevant entries for determining party wins

````python
filtered_elections_df = elections_df[~elections_df['candidate'].isin(['OVERVOTES', 'UNDERVOTES'])]
filtered_elections_df = filtered_elections_df[filtered_elections_df['party_detailed'] != 'Unknown']
````

Group by year, state, and party_detailed to sum candidate votes for each party in each state and year

````python
party_votes_by_year_state = filtered_elections_df.groupby(['year', 'state', 'party_detailed'])['candidatevotes'].sum().reset_index()
````

Find the party with the maximum votes for each year and state, This identifies the winner of each presidential election in each state

````python
winning_party_per_election = party_votes_by_year_state.loc[party_votes_by_year_state.groupby(['year', 'state'])['candidatevotes'].idxmax()]
````

Count the number of wins for each party in each state across all years

````python
state_party_wins = winning_party_per_election.groupby(['state', 'party_detailed']).size().reset_index(name='wins')
````

For each state, find the party with the most wins

````python
most_frequent_winner_per_state = state_party_wins.loc[state_party_wins.groupby('state')['wins'].idxmax()]
````

Display the results

````python
display(most_frequent_winner_per_state.sort_values(by='state').reset_index(drop=True))
````

### Answer 3.

Create a new column to categorize parties as 'Major' or 'Other'

````python
def categorize_party(party):
    if party in ['DEMOCRAT', 'REPUBLICAN']:
        return 'Major'
    else:
        return 'Other'

elections_df['party_category'] = elections_df['party_detailed'].apply(categorize_party)
````

Calculate the average candidate votes for each party category

````python
average_votes_by_party_category = elections_df.groupby('party_category')['candidatevotes'].mean().reset_index()
````

Display the results

````python
display(average_votes_by_party_category)
````

Visualize the results

````python
plt.figure(figsize=(8, 5))
sns.barplot(data=average_votes_by_party_category, x='party_category', y='candidatevotes', palette='viridis')
plt.title('Average Candidate Votes by Party Category')
plt.xlabel('Party Category')
plt.ylabel('Average Candidate Votes');
````

### Answer 4.

Calculate total 'OTHER' party votes per state per year

````python
other_votes_by_state_year = elections_df[elections_df['party_category'] == 'Other'].groupby(['year', 'state'])['candidatevotes'].sum().reset_index()
other_votes_by_state_year.rename(columns={'candidatevotes': 'other_party_votes'}, inplace=True)
````

Calculate total votes per state per year

````python
total_votes_by_state_year = elections_df.groupby(['year', 'state'])['totalvotes'].sum().reset_index()
total_votes_by_state_year.drop_duplicates(subset=['year', 'state'], inplace=True)
````

Merge to get proportions

````python
state_proportions = pd.merge(other_votes_by_state_year, total_votes_by_state_year, on=['year', 'state'], how='left')
state_proportions['other_proportion'] = state_proportions['other_party_votes'] / state_proportions['totalvotes']
````

Calculate national total votes and 'OTHER' party votes per year

````python
national_total_votes_by_year = elections_df.groupby('year')['totalvotes'].sum().reset_index()
national_total_votes_by_year.drop_duplicates(subset=['year'], inplace=True)
national_other_votes_by_year = elections_df[elections_df['party_category'] == 'Other'].groupby('year')['candidatevotes'].sum().reset_index()
national_other_votes_by_year.rename(columns={'candidatevotes': 'national_other_votes'}, inplace=True)
````

Calculate national 'OTHER' party proportion

````python
national_proportions = pd.merge(national_other_votes_by_year, national_total_votes_by_year, on='year', how='left')
national_proportions['national_other_proportion'] = national_proportions['national_other_votes'] / national_proportions['totalvotes']
````

Merge state and national proportions

````python
comparison_df = pd.merge(state_proportions, national_proportions[['year', 'national_other_proportion']], on='year', how='left')
````

Identify states where 'OTHER' party proportion is higher than the national average

````python
comparison_df['is_higher_than_national_avg'] = comparison_df['other_proportion'] > comparison_df['national_other_proportion']
````

Filter for states that are consistently higher (e.g., in more than 50% of the elections they participated in)

````python
consistent_higher_states = comparison_df.groupby('state')['is_higher_than_national_avg'].mean().reset_index()
consistent_higher_states = consistent_higher_states[consistent_higher_states['is_higher_than_national_avg'] > 0.5]
````

Disply the result

````python
display(consistent_higher_states.sort_values(by='is_higher_than_national_avg', ascending=False))
````

### Answer 5.

Aggregate candidate votes by year and party for major parties

````python
major_party_votes_trend = elections_df[elections_df['party_detailed'].isin(['DEMOCRAT', 'REPUBLICAN'])]
major_party_votes_trend = major_party_votes_trend.groupby(['year', 'party_detailed'])['candidatevotes'].sum().reset_index()
````

Plotting the trend for major parties

````python
plt.figure(figsize=(14, 7))
sns.lineplot(data=major_party_votes_trend, x='year', y='candidatevotes', hue='party_detailed', marker='o')
plt.title('Candidate Votes Trend for Major Parties Over Election Years')
plt.xlabel('Election Year')
plt.ylabel('Total Candidate Votes')
plt.grid(True)
plt.xticks(major_party_votes_trend['year'].unique(), rotation=45)
plt.legend(title='Party');
````

To explore other parties, we can look at the top 'Other' parties

````python
top_other_parties = elections_df[elections_df['party_category'] == 'Other']['party_detailed'].value_counts().nlargest(5).index
other_party_votes_trend = elections_df[elections_df['party_detailed'].isin(top_other_parties)]
other_party_votes_trend = other_party_votes_trend.groupby(['year', 'party_detailed'])['candidatevotes'].sum().reset_index()
````

````python
if not other_party_votes_trend.empty:
    plt.figure(figsize=(14, 7))
    sns.lineplot(data=other_party_votes_trend, x='year', y='candidatevotes', hue='party_detailed', marker='o')
    plt.title('Candidate Votes Trend for Top Other Parties Over Election Years')
    plt.xlabel('Election Year')
    plt.ylabel('Total Candidate Votes')
    plt.grid(True)
    plt.xticks(other_party_votes_trend['year'].unique(), rotation=45)
    plt.legend(title='Party', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
else:
    print("No significant 'Other' party data to visualize trends.")
````
