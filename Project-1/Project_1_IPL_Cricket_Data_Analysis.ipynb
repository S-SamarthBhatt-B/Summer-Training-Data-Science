# **IPL Cricket Data Analysis**

## **Objective:**
### Analyze IPL matches and player statistics.

## **Dataset:**
### IPL datasets from Kaggle.

## **Features:**
•	Top run scorers

•	Most wickets

•	Team winning percentage

•	Toss impact analysis

## **Concepts:**

•	GroupBy

•	Merge

•	Pivot Tables

•	NumPy Calculations

```python
import pandas as pd
import numpy as np
```

```python
ipl_df=pd.read_csv('/content/IPL.csv')
```

```python
ipl_df
```

# 1. Top Run Scorers

```python

top_scorers = ipl_df.groupby('batter')['batter_runs'].sum().sort_values(ascending=False).head(10)
```

## --- Top 10 Run Scorers ---


```python
print(top_scorers)
```

# 2. Most Wickets (Filtering for deliveries that resulted in a bowler wicket)

```python
most_wickets = ipl_df[ipl_df['bowler_wicket'] == 1].groupby('bowler').size().sort_values(ascending=False).head(10)
```

##--- Top 10 Wicket Takers ---

```python
print(most_wickets)
```

# 3. Team Winning Percentage

```python
match_winners = ipl_df.groupby('match_id')['match_won_by'].first()
total_matches = ipl_df.groupby('batting_team')['match_id'].nunique()
wins = match_winners.value_counts()
win_percentage = (wins / total_matches * 100).dropna().sort_values(ascending=False)
```

##--- Team Win Percentage ---


```python
print(win_percentage)
```

# 4. Toss Impact Analysis

```python
toss_match_winners = ipl_df.groupby('match_id')[['toss_winner', 'match_won_by']].first()
toss_impact = (toss_match_winners['toss_winner'] == toss_match_winners['match_won_by']).value_counts(normalize=True) * 100
```

##--- Toss Impact (Win Toss & Win Match %) ---

```python
print(toss_impact)
```
